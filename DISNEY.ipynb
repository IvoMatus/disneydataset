```python
from bs4 import BeautifulSoup as bs
from bs4 import NavigableString, Tag
import requests
```

```python
#Cargar la página
r = requests.get("https://es.wikipedia.org/wiki/Toy_Story_3")

#Convertir a objeto bs4
soup = bs(r.content)
#imprimir

contents = soup.prettify()
print(contents)
```

```python
#obtener la info necesaria
info_box = soup.find(class_="infobox plainlist plainlinks")
#print(info_box.prettify())
info_rows = info_box.find_all("tr")
for row in info_rows:
    print(row.prettify())
```

```python
movie_info = {}

for index, row in enumerate(info_rows):
    if index == 0:
        movie_info['Titulo'] = row.find("th").get_text()
    elif index == 1 or index == 2 or index == 13 or index == 22:
        continue # porque es la foto de la película o cabecera
    #elif (2 < index < 12) or (13 < index < 16) or (22 < index < 27):
    elif (2 < index < 15) or (16 < index <24) or (24 < index < 30):
        content_key = row.find("th").get_text()
        content_value = row.find("td").get_text().replace("\n"," ").replace('200b','')
        #content_value = get_content_value(row.find("td"))
        movie_info[content_key] = content_value

movie_info
```

```python

def get_content_value(row_data): # hay casos en que hay break lines entonces:
    if row_data.find("a"):
        if len(row_data.find_all("a"))> 1:
            
            return [li.get_text() for li in row_data.find_all("a")]
        else:
            return row_data.get_text().replace("\n","").replace("[2]\u200b","")
#        return [li.get_text() for li in row_data.find_all("ul")]
    else: # si la fila no tiene lista entonces vale pico
         return row_data.get_text().replace("\n","")
```

```python
movie_info={}
for index, row in enumerate(info_rows):
    
    if index == 0:
        movie_info['Title'] = row.find("th").get_text()
    elif index == 1 or index == 2 or index == 13 or index == 22:
        continue # porque es la foto de la película o cabeceras de tabla zZz
    elif (2 < index < 15) or (16 < index <24) or (24 < index < 30):
        content_key = row.find("th").get_text()
        content_value = get_content_value(row.find("td"))
        movie_info[content_key] = content_value
if movie_info['Estreno']:
    del movie_info['Estreno']
movie_info
```
